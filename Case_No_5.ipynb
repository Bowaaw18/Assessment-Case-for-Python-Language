```python
case_5 = pd.read_csv("Documents/Data_Science/dataset_superstore_simple.csv", parse_dates = ["order_date"])
case_5
```

```python
case_5["order_year"] = case_5.order_date.dt.strftime("%Y")
case_5
```

```python
step1 = case_5.groupby(["order_year"]).agg({"sales":"sum", "profit":"sum"}).reset_index()
step1
```

```python
step2 = case_5[["order_year","customer_id"]]
step2
```

```python
step3 = step2.groupby(["order_year"]).count()
step3
```

```python
result_No_5 = pd.merge (step1,step3, how = "left", on = "order_year")
result_No_5
```
